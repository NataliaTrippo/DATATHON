```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure df_plot is correctly prepared with 'Ano ingresso' column
df_plot = dados_filtrados_por_ano.groupby('Ano ingresso')['Defas'].value_counts().reset_index(name='Count')

plt.figure(figsize=(12, 7))
ax = sns.barplot(data=df_plot, x='Defas', y='Count', hue='Ano ingresso', palette='viridis')
plt.title('Total de Alunos em Defasagem (2020, 2021, 2022)')
plt.xlabel('Valor de Defasagem')
plt.ylabel('') # Remove y-axis label
plt.xticks(rotation=45)
plt.yticks([]) # Remove y-axis numbers
plt.legend(title='Ano de Admissão')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add total count on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline',
                xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

___

O ano de 2022 apresenta o maior número de alunos em cada categoria 'Defas', indicando um crescimento no número total de alunos admitidos ao longo desses anos.

* **Alunos na Fase Ideal ('Defas 0')**: Um número considerável de alunos encontra-se na sua fase acadêmica ideal ('Defas 0'), particularmente em 2022, embora esse número seja menor do que o de alunos com valores negativos de 'Defas'. * **Poucos Alunos Adiantados ('Defas 1')**: Constatou-se que muito poucos alunos estão adiantados em relação à sua fase acadêmica ideal ('Defas 1') em qualquer um dos anos de admissão, sugerindo que estar adiantado é raro.

# **DESEMPENHO ACADÊMICO**

```python
import pandas as pd
import openpyxl
dados = pd.read_excel('/content/BASE DE DADOS PEDE 2024 - DATATHON.xlsx')

import matplotlib.pyplot as plt
import seaborn as sns

# Filter for the years of interest directly from the 'dados' DataFrame
anos_interesse = [2020, 2021, 2022]
df_ida_filtered = dados[dados['Ano ingresso'].isin(anos_interesse)].copy()

# Calculate the mean of 'IDA' grouped by 'Ano ingresso'
df_media_ida = df_ida_filtered.groupby('Ano ingresso')['IDA'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
ax = sns.lineplot(data=df_media_ida, x='Ano ingresso', y='IDA', marker='o')

plt.title('Evolução do Desempenho Acadêmico Médio (IDA)')
plt.xlabel('Ano de Ingresso')
plt.ylabel('Média do IDA')
plt.xticks(anos_interesse) # Set x-axis ticks to only the years of interest
plt.grid(True)

# Highlight critical points (min and max IDA)
min_ida_row = df_media_ida.loc[df_media_ida['IDA'].idxmin()]
max_ida_row = df_media_ida.loc[df_media_ida['IDA'].idxmax()]

# Plot min IDA point
ax.plot(min_ida_row['Ano ingresso'], min_ida_row['IDA'], 'o', color='red', markersize=10, label=f'Mínimo: {min_ida_row['IDA']:.2f}')
ax.annotate(f'Min: {min_ida_row['IDA']:.2f}',
            (min_ida_row['Ano ingresso'], min_ida_row['IDA']),
            textcoords="offset points", xytext=(0,10), ha='center', color='red', fontsize=10)

# Plot max IDA point
ax.plot(max_ida_row['Ano ingresso'], max_ida_row['IDA'], 'o', color='green', markersize=10, label=f'Máximo: {max_ida_row['IDA']:.2f}')
ax.annotate(f'Max: {max_ida_row['IDA']:.2f}',
            (max_ida_row['Ano ingresso'], max_ida_row['IDA']),
            textcoords="offset points", xytext=(0,-15), ha='center', color='green', fontsize=10)

plt.tight_layout()
plt.show()
```

**Resumo das informações do gráfico de evolução do IDA (2020–2022)**
___

O gráfico de linha mostra a evolução da média do Índice de Desempenho Acadêmico (IDA) ao longo dos três anos de ingresso:

* Estabilidade inicial (2020–2021): Em 2020, a média do IDA ficou ligeiramente acima de 6,5 e atingiu seu ponto máximo em 2021 (6,53). Esse resultado sugere um período de estabilidade e até leve melhora no desempenho acadêmico médio.

* Queda significativa em 2022: No ano de 2022, a média caiu para 6,07, o menor valor da série. Essa redução indica uma piora no desempenho acadêmico médio dos estudantes ingressantes.

* Tendência geral: Apesar da estabilidade entre 2020 e 2021, o declínio em 2022 chama atenção e pode estar relacionado a fatores externos (como mudanças no perfil dos alunos, contexto socioeconômico ou adaptações institucionais) que impactaram negativamente o desempenho.

Em conclusão, o desempenho acadêmico médio (IDA) manteve-se estável entre 2020 e 2021, com leve melhora, mas apresentou queda acentuada em 2022, indicando uma tendência geral de declínio após um período inicial de estagnação. Esse resultado reforça a necessidade de investigar os fatores que contribuíram para essa redução e de implementar estratégias de apoio para recuperar o desempenho nos anos seguintes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Filter for the years 2020, 2021, 2022 and prepare metrics
anos_interesse = [2020, 2021, 2022]
df_filtered = dados[dados['Ano ingresso'].isin(anos_interesse)]
df_rel = df_filtered.groupby('Ano ingresso')[['IEG', 'IDA', 'IPV']].mean().reset_index()

# Calculate correlation matrix
corr = df_rel[['IEG', 'IDA', 'IPV']].corr()

# Custom palette implementation
# We add a dark gray (#404040) at the very end of the palette specifically for 1.00
custom_colors = ["#52423c", "#ad5c70", "#d3ad98", "#edd4be", "#b9c3c4", "#404040"]
custom_cmap = sns.blend_palette(custom_colors, as_cmap=True)

# Plot heatmap with the custom palette
plt.figure(figsize=(8, 6))
# We set vmax slightly below 1 to ensure 1.00 hits the last color (dark gray)
sns.heatmap(corr, annot=True, cmap=custom_cmap, fmt=".2f", vmin=-1, vmax=1.001, center=0)
plt.title('Correlação IDA, IEG e IPV')
plt.show()
```

**Resumo do gráfico de correlação entre IEG, IDA e IPV**
___

O gráfico mostra como as três variáveis se relacionam entre si:

 * IDA e IPV: Existe uma correlação forte e positiva (0,94), indicando que quando o desempenho acadêmico (IDA) aumenta, o índice de permanência (IPV) também tende a aumentar. Isso sugere que alunos com melhor desempenho permanecem mais na instituição.

 * IEG e IPV: Há uma correlação moderada e negativa (-0,51), mostrando que valores mais altos de IEG estão associados a menores índices de permanência.

 * IEG e IDA: A relação é fraca e negativa (-0,18), sugerindo pouca influência direta entre o índice de ingresso (IEG) e o desempenho acadêmico médio (IDA).

Em conclusão, o grau de engajamento dos alunos (IEG) não apresenta relação direta forte com o desempenho (IDA) nem com o ponto de virada (IPV), já que suas correlações são fracas ou moderadamente negativas. Em contrapartida, IDA e IPV estão fortemente associados, indicando que o desempenho acadêmico é o principal fator ligado à permanência, mais do que o engajamento inicial.

# **ASPECTOS PSICOPEDAGÓGICOS**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Recriar a categoria IAN (necessário pois a célula anterior foi removida)
def categorizar_ian(valor):
    if pd.isna(valor): return 'Indefinido'
    if valor < 5: return 'Baixo'
    elif valor < 7: return 'Médio'
    else: return 'Alto'

dados['IAN_categoria'] = dados['IAN'].apply(categorizar_ian)

plt.figure(figsize=(10, 6))

# 2. Gráfico de Regressão (IAN vs IPV)
sns.regplot(
    data=dados,
    x='IAN',
    y='IPV',
    scatter_kws={'color': '#b9c3c4'},
    line_kws={'color': '#52423c'}
)

# 3. Boxplot sobreposto (Ajustando paleta para 3 categorias)
sns.boxplot(
    data=dados,
    x='IAN_categoria',
    y='IPV',
    palette=['#52423c', '#ad5c70', '#d3ad98'],
    hue='IAN_categoria',
    legend=False
)

plt.title('Relação entre IAN (Adequação) e IPV (Ponto de Virada)', color='#404040')
plt.xlabel('IAN / Categoria', color='#404040')
plt.ylabel('IPV', color='#404040')

plt.grid(alpha=0.3)
plt.show()
```

**Resumo do gráfico de relação entre IAN (Adequação) e IPV (Ponto de Virada)**
___

*   Categoria “Alto”: Apresenta a maior mediana de IPV, indicando que alunos com alta adequação psicopedagógica tendem a ter maior permanência e estabilidade acadêmica.

*   Categoria “Médio”: Mostra valores intermediários de IPV, sugerindo que a adequação moderada ainda contribui positivamente, mas com menor impacto.

*   Categoria “Baixo”: Apresenta a menor mediana de IPV, confirmando que baixa adequação psicopedagógica está associada a menor permanência e maior risco de evasão.

 * Presença de outliers: Em todas as categorias há casos fora do padrão, mas a tendência geral se mantém clara: quanto maior a adequação, maior o IPV.

As avaliações psicopedagógicas (IPP) confirmam a defasagem identificada pelo IAN, já que níveis mais baixos de adequação se relacionam diretamente com menor permanência (IPV). Isso reforça que o aspecto psicopedagógico é um fator determinante para explicar e validar os padrões de defasagem acadêmica.

# **MULTIDIMENSIONALIDADE**

```python
import pandas as pd

# Ajustando para os nomes de colunas reais identificados no DataFrame
# Nota: 'INDE 22' é a coluna de índice, e usamos os indicadores sem o sufixo _2022
df = dados[[
    'INDE 22',
    'IDA',
    'IEG',
    'IPS',
    'IAN'
]].dropna()
```

```python
from sklearn.linear_model import LinearRegression

# Definindo as variáveis independentes e a dependente (INDE)
X = df[['IDA','IEG','IPS','IAN']]
y = df['INDE 22']

modelo = LinearRegression()
modelo.fit(X, y)

coeficientes = pd.Series(modelo.coef_, index=X.columns).sort_values(ascending=False)
print(coeficientes)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.barplot(
    x=coeficientes.values,
    y=coeficientes.index,
    palette=['#52423c','#ad5c70','#d3ad98','#edd4be','#b9c3c4','#404040']
)

plt.title('Influência dos Indicadores no INDE', color='#404040')
plt.xlabel('Peso (Influência)')
plt.ylabel('Indicadores')

plt.grid(axis='x', alpha=0.3)
plt.show()
```

**Resumo do gráfico de influência dos indicadores no INDE**
____

*  IEG (Engajamento – ~0,29): É o fator mais influente. Isso significa que o nível de participação e envolvimento dos alunos nas atividades acadêmicas e institucionais tem o maior impacto sobre a nota global. Em termos práticos, alunos que se engajam mais tendem a elevar significativamente seu INDE, mesmo quando outros indicadores não são tão altos.

*   IDA (Desempenho Acadêmico – ~0,25): Vem logo em seguida, mostrando que o desempenho acadêmico é quase tão determinante quanto o engajamento. A combinação de alto engajamento + bom desempenho é a mais poderosa para elevar a nota global.
    
*  IPS (Aspectos Psicossociais – ~0,14): Tem influência moderada. Isso indica que fatores emocionais e sociais funcionam como suporte: não são os principais motores da nota, mas ajudam a sustentar o desempenho e o engajamento.

*  IAN (Adequação – ~0,10): É o indicador com menor peso. Embora contribua, sua influência é limitada, funcionando mais como complemento do que como fator decisivo para o INDE.

**Conclusão**
____

A análise da multidimensionalidade do INDE mostra que a nota global do aluno não depende de um único indicador isolado, mas sim da combinação equilibrada entre diferentes dimensões:

O engajamento (IEG) e o desempenho acadêmico (IDA) aparecem como os pilares centrais, exercendo maior influência sobre o resultado final.

Já os aspectos psicossociais (IPS) atuam como reforço, contribuindo para a consistência e estabilidade do desempenho e da participação, enquanto o IAN desempenha um papel secundário, com impacto mais limitado.

O INDE se eleva de forma mais significativa quando há uma **integração entre alto engajamento e bom desempenho acadêmico**, apoiados por condições psicossociais positivas, mostrando que o resultado global vem da interação entre comportamento, desempenho e contexto.

# **EFETIVIDADE DO PROGRAMA**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Organizar os dados existentes para o formato longo
df_long = pd.DataFrame({
    'Ano': ['2020']*len(dados) + ['2021']*len(dados) + ['2022']*len(dados),
    'PEDRA': pd.concat([dados['Pedra 20'], dados['Pedra 21'], dados['Pedra 22']]),
    'INDE': pd.concat([dados['INDE 22'], dados['INDE 22'], dados['INDE 22']])
}).dropna(subset=['PEDRA'])

# Média por grupo de Ano e Pedra
media = df_long.groupby(['Ano', 'PEDRA'])['INDE'].mean().reset_index()

# Plot de barras
plt.figure(figsize=(14, 7))

ax = sns.barplot(
    data=media,
    x='Ano',
    y='INDE',
    hue='PEDRA',
    palette='viridis'
)

# Adicionar os valores no topo das colunas
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'{p.get_height():.2f}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='baseline',
                    fontsize=9, color='black', xytext=(0, 5),
                    textcoords='offset points')

# Criar legenda personalizada com cores da paleta viridis
pedras_labels = sorted(media['PEDRA'].unique())
colors = sns.color_palette('viridis', len(pedras_labels))
legend_patches = [mpatches.Patch(color=colors[i], label=pedras_labels[i]) for i in range(len(pedras_labels))]

# Mover a legenda para fora (lado direito)
plt.legend(handles=legend_patches, title='Fases (Pedras)', loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0, frameon=True)

plt.title('Média do INDE por Fase (PEDRA) e Ano', color='#404040', loc='left')
plt.xlabel('Ano da Classificação')
plt.ylabel('') # Remove o rótulo do eixo Y
plt.yticks([]) # Remove os números/escala do eixo Y

plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Resumo do gráfico de média INDE por Fase e Ano**
___

O gráfico mostra a evolução da média do INDE em cada fase (Quartzo, Ágata, Ametista e Topázio) ao longo dos anos de ingresso (2020, 2021 e 2022):

  * Topázio: Apresenta os maiores valores de INDE e uma trajetória de crescimento consistente (7,78 → 8,09 → 8,37). Isso indica que nesta fase o programa tem impacto muito positivo e contínuo.

  * Ametista: Também mostra melhora progressiva (7,07 → 7,14 → 7,53), confirmando ganhos ao longo do ciclo.

  * Ágata: Evolui de forma estável e crescente (6,19 → 6,50 → 6,61), sugerindo que o programa contribui para avanços graduais.

  * Quartzo: É a única fase que não apresenta melhora consistente. Apesar de um aumento em 2021 (5,58 → 5,88), cai em 2022 (5,24), indicando fragilidade ou menor impacto do programa nessa fase.

  A fase Quartzo é a única que não apresenta melhora consistente ao longo do ciclo.
```python
import pandas as pd

df = dados.copy()
ian_data = df['IAN']
ida_data = df['IDA']
```

```python
features = [
    'IDA',
    'IEG',
    'IPS',
    'IAA',
    'IAN'
]

X = df[features]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


y_binary = (df['INDE 22'] < 6.0).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

modelo = LogisticRegression(max_iter=1000)
modelo.fit(X_train, y_train)
```

```python
df['prob_risco'] = modelo.predict_proba(X)[:,1]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['prob_risco'], bins=20, color='#ad5c70')
plt.title('Distribuição de Risco dos Alunos')
plt.show()
```

```python
def classificar_risco(p):
    if p > 0.7:
        return 'Alto risco'
    elif p > 0.4:
        return 'Médio risco'
    else:
        return 'Baixo risco'

df['nivel_risco'] = df['prob_risco'].apply(classificar_risco)
```
